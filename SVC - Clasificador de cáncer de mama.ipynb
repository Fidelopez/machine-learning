# Preparación de datos
###### Pasos a realizar:
- Dividir datos en sets de entrenamiento y testing
- Normalizar datos con valores entre 0 y 1

```python
X = cancer_df.drop('target', axis=1)
y = cancer_df['target']
```

```python
from sklearn.model_selection import train_test_split

# print(help(train_test_split))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_min = np.min(X_train)
X_train_scaled = (X_train - X_min) / (np.max(X_train) - X_min)
X_train_scaled.head()
```

```python
X_test_min = np.min(X_test)
X_test_scaled = (X_test - X_test_min) / (np.max(X_test) - X_test_min)
X_test_scaled.head()
```

# Entrenamiento y optimización del modelo SVC
###### Pasos a seguir:
- Importar librerías necesarias
- Realizar un GridSearchCV para encontrar los mejores parámetros para el modelo
- Entrenar el modelo y predecir con el set de prueba
- Visualizar resultados

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
grid_params = {'C': [0.1, .5, 1, 5, 10,100], 'gamma':[0.1, .5,1,2,5,10,100], 'kernel':['rbf']}
grid = GridSearchCV(SVC(), grid_params, cv=5)
```

```python
grid.fit(X_train_scaled, y_train)
print(grid.best_params_)
print(grid.best_score_)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib
y_predict = grid.predict(X_test_scaled)
matrix = confusion_matrix(y_test, y_predict)
akws = {"ha": 'center',"va": 'center'}
ax = sns.heatmap(matrix, annot=True, fmt="d", annot_kws=akws)
```

```python
print(classification_report(y_test, y_predict))
```
